***Objective 1:***

 To visualize the distribution of used car prices across different fuel types and transmission types, and how these distributions evolve over time.

```python
# Animated violin plot for Price distribution by Fuel Type and Transmission
fig_price_distribution = px.violin(df, x='Fuel Type', y='Price', color='Transmission',
                                   animation_frame='Year',
                                   title='Price Distribution by Fuel Type and Transmission',
                                   labels={'Fuel Type': 'Fuel Type', 'Price': 'Price'},
                                   category_orders={'Transmission': ['Automatic', 'Manual']})

# Show the animated violin plot
fig_price_distribution.show()
```

***Analysis:***

Wider violins indicate broader price ranges, while narrower violins suggest more concentrated prices.
Skewed violins point to price distributions that are not symmetrical.

**Fuel Type and Price:**
Relationships between fuel type and price can be observed (e.g., one fuel type might generally have higher prices than others).

**Transmission and Price:**
Differences in price distributions for automatic and manual cars within each fuel type can be identified.

**Yearly Trends:**
The animation shows how price distributions for different fuel types and transmissions have shifted over time, reflecting market dynamics and technology advancements.

**Objective 3:**

To create an animated line plot that visualizes the yearly trends of average mileage (kilometers) for different engine types, allowing for comparisons between them and identification of patterns over time.

```python
# Sort the engine types alphabetically
sorted_engine_types = sorted(df['Engine'].unique())

# Calculate the yearly average price and average kilometer for each engine type
avg_price_per_engine = df.groupby(['Year', 'Engine'])['Price'].mean().reset_index()
avg_kilometer_per_engine = df.groupby(['Year', 'Engine'])['Kilometer'].mean().reset_index()
# Animated line plot for Yearly trend of Average Kilometer by Engine
fig_avg_kilometer = px.line(avg_kilometer_per_engine, x='Year', y='Kilometer', color='Engine',
                             title='Yearly Trend of Average Kilometer by Engine',
                             labels={'Year': 'Year', 'Kilometer': 'Average Kilometer'},
                             animation_frame='Engine')
fig_avg_kilometer.for_each_trace(lambda t: t.update(name=sorted_engine_types.index(t.name)))

fig_avg_kilometer.show()
```

***Analysis:***

**General Price Trends:** Average prices for most engine types have increased over time, with some fluctuations.

**Engine-Specific Trends:** The animation enables observing distinct

**trends for each engine type, revealing:**
Engine types with consistently higher or lower average prices.
Engine types with significant price increases or decreases over time.
Engine types that exhibit more volatile price patterns.

**Identifying Outliers:** Points that deviate significantly from overall trends can be explored for potential insights.

***Further Insights:***

**Correlations:** Explore relationships between engine type, price, mileage, year, fuel type, vehicle segment, or other variables.
